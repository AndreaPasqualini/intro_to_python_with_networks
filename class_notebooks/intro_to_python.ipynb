## The Basics

Python alone cannot do much.
For this reason, we are almost always going to work with a package (see below).
However, it is fundamental to understand the basics.
This involves familiarizing with the _syntax_ and with the basic _data types_.

Syntax is the set of rules that govern writing code.
This includes how to write an _assignment_ (providing a variable with a value), how to call functions and how to access items in a iterable object (e.g., lists, arrays).
It also includes _code blocks,_ which execute conditionally on a given rule (e.g., `if`, `while`).

Data types are the types of variables we can create.
Python defines a few basic ones and packages (see below) provide new data types.
This programming language features _dynamic typing_, which means that we do not have to define what a variable can be.
Instead, Python infers the type of a variable when we create it.
Examples of basic data types are strings (`str`), lists (`list`), integer numbers (`int`).

As with many programming languages, we are defining variables and changing their values all the time in Python.
We can create a variable simply by inventing a name and assigning a value to it, like the following.

```python
a = 1
```

Now we have a variable whose name is `a` and its value is `1`.
This variable has a _type_, which is the kind of value it contains.
In particular, `1` is an integer, whose type is represented in Python with the keyword `int`.
Understanding that variables can be of different types is important, because the type defines what we can do with that variable.
We can ask Python to tell us the type of a variable by using the `type()` function.

```python
type(a)
```

### Control flow (`if`, `for` and `while`)

Control flow is the set of techniques that allows us to manage the way code is executed.
For example, you might want to execute some code provided a certain condition is `True`.
Or you might want to repeat a certain action a given number of times.
Or, again, you might want to run code as long as a condition is `True`, but you do not know how many times the code should be executed before the condition turns `True`.

The three examples are managed by the statements [`if`](https://docs.python.org/3/reference/compound_stmts.html#the-if-statement), [`for`](https://docs.python.org/3/reference/compound_stmts.html#the-for-statement) and [`while`](https://docs.python.org/3/reference/compound_stmts.html#the-while-statement).
Here we see basic applications, just to convey the idea.
We should keep in mind two things when working with control flow statements:
- The control flow statement has to be on a line that end with a semicolon (`:`).
- The code block following the control flow statements has to be indented with four whitespaces.

The `if` and `while` statements work provided that a certain condition holds.
Such condition is a boolean value that is either `True` or `False`.
Python supports boolean arithmetics with the keywords `and`, `or` and `not`.

The `if` statement can be expanded to the `if`-`elif`-`else`.

```python
condition = True
```

```python
if condition:
    print("The condition is true.")
else:
    print("The condition is false.")
```

```python
if not condition:
    print("The condition is false.")
else:
    print("The condition is true.")
```

```python
cond1 = 2 == 2.0
cond2 = 3 <= 0
```

```python
if cond1 and cond2:
    print("The two conditions are both true.")
else:
    print("At least one condition is false.")
```

```python
if cond1 or cond2:
    print("At least one condition is true.")
else: 
    print("Both conditions are false.")
```

```python
if cond1 and cond2:
    print("Both conditions are true.")
elif cond1:
    print("Only condition 1 is true.")
elif cond2:
    print("Only condition 2 is true.")
else:
    print("Neither condition is true.")
```

We now turn to the `for` statement.
This is useful when we know that we want to run a code block a given number of times.
For example, given the `phonebook` object before, suppose we want to `print` all the names.

```python
for entry in phonebook:
    print(entry[0])
```

A `for` loop always _walks an iterable_.
This means that we will always see the syntax **`for`** `element` **`in`** {`list`, `tuple`, `dict`, other iterable}.
This is the case even with numbers.
Suppose we want to print all integer numbers up to 5 (excluded).

```python
for i in range(5):
    print(i)
```

The function [`range()`](https://docs.python.org/3/library/functions.html#func-range) lists all integer numbers starting from zero up to 5, excluding 5.

Finally, the `while` statement executes code up until a certain condition is verified.
A fun way to use it is to approximate the Euler constant $e$.
We use the following definition of $e$:

$$ e := \lim_{n \to \infty} {\left( 1 + \dfrac{1}{n} \right)}^n $$

We cannot numerically compute this limit.
What we can do, though, is to compute the argument of the limit for given values $n-1$ and $n$.
We verify whether we significantly improve the result between the $(n-1)$-th and the $n$-th iteration (note that we are essentially applying the $(\varepsilon, \delta)$ definition of a limit).
If the result does not change much in two successive iterations, then we got to a "good enough" approximation of $e$.
Otherwise, we increase $n$.

We cannot know ex-ante how much $n$ should be to have a "good enough" approximation, and so we need to try to compute $e$ _until_ we get a nice result.

```python
# Starting conditions, just to get the loop started
improvement = 1.0
tolerance = 1e-10  # 10^(-10)
n = 1
step_increase = 1
results = [1.0]

# Here is the iterative algorithm to obtain e
while improvement > tolerance:           # as long as we improve significantly
    tentative = ( 1 + (1 / n) ) ** n     # compute a tentative result
    results.append( tentative )          # store result in a list
    improvement = abs( results[-1] - results[-2] )  # check the improvement
    n += step_increase                   # increase this value
    
# Print results to obtain information about the algorithm
print("This code ran " + str(n) + " times.")
print("The approximated Euler's constant is " + str(results[-1]) + ".")
```

Again, note that we could not have known that we would have needed around that many iterations.
As a fun exercise, try to change the `tolerance` or the `step_increase` to see how the approximation and the speed of convergence change.

If you look close enough, you will see the syntax `results.append()`.
This is a _method_ common to all lists, which appends an item in the last position of the list.
We will briefly cover methods below.

Finally, we can even define what happens when a certain error occurs.
This can happen with the [`try`](https://docs.python.org/3/reference/compound_stmts.html#the-try-statement)-`catch` statements, but this goes outside the scope of this TA session.

### Functions (`def` and `lambda`)

You might have noticed so far that Python is rather limited as a programming language.
This is because Python is a general-purpose language: it should be flexible enough for everybody to achieve their goals, but it cannot afford to be a BIG library of code.

For example, Python does not know what a square root of a number is.
While we can always rely on the fact that $\sqrt[n]{x} = x^{1/n}$, it might be useful to define a dedicated function that does the job.
In this example, defining a function improves code readability.

We can define a new function by using the `def` keyword, listing arguments in round parentheses.
Also, if we want the function to give back something, we should explicitly instruct Python to do so with the `return` keyword.

```python
def sqrt(x):
    return x ** (1/2)
```

Here we defined the square-root function.
As it is, this function works for all values of $x$, even negative ones (the result will be a complex number).

Again, note that the `def` statement ends with a semicolon (`:`) and any code that belongs to the function is indented with respect to the line containing the keyword `def`.

```python
sqrt(2)
```

An example of a slightly more sophisticated function is one that computes the real roots a (non-negative) number.

```python
def real_nth_root(x, n=2):
    if x < 0:
        raise ValueError("Root argument must be non-negative.")
    return x ** (1/n)
```

This function performs the so-called _input validation_.
To ensure that the result of $\sqrt[n]{x}$ is a real number, we must ensure that $x$ is non-negative.
We do this with an `if` clause and instructing Python to terminate the execution of the function with a `raise` command, which issues an error.
Also, the function we defined takes two arguments: `x` and `n`.
By specifying `n=2` in the `def` statement, we are saying that the default value of $n$ should be two, but the user can change that value arbitrarily.

```python
real_nth_root(2)  # we expect the square root, because we omitted n
```

```python
real_nth_root(-2)
```

```python
real_nth_root(125, 3)
```

Finally, a quick-and-dirty way to define functions is by using `lambda` functions, which are also known as [anonymous functions](https://en.wikipedia.org/wiki/Anonymous_function).
The inner workings of anonymous functions are quite complicated and we will not cover them.
Here it should suffice to know that these functions are defined _in-line_, which means that we do not write a code block for them.
Anonymous functions are useful for quick one-liners that do not require much work.

We can re-define the function `sqrt` above as an anonymous function.

```python
root2 = lambda x: x ** (1/2)
```

The syntax is as follows.
The function `root2` takes an argument `x`, which is indicated right after the keyword `lambda`.
There is a semicolon following the statment, after which we find the main task performed by the function we defined.

```python
root2(2)
```

### Methods

Methods are nothing else than functions.
However, methods are defined for specific objects.
For example, we can easily understand that we might want to attach an item at the end of _any_ `list`.
Or we can see that for _any_ `complex` number, there is a real and an imaginary part to it.
Methods are functions that come "attached" to objects of a given type.

Methods are accessed with the syntax `object.method`.
We can have two main types of methods: properties and functions.
Functions take arguments with the usual round parentheses (e.g., `list.append(item)`).
Properties do not require parentheses (e.g., `complex.imag`).

Remember that we defined above the variable `c` containing a complex number.

```python
c
```

Any `complex` object comes with two methods.
One is `real` and the other is `imag`.
No explanation is required.

```python
c.real
```

```python
c.imag
```

Similarly, any `list` comes with a `append` method, which, as we saw above, takes an argument and attaches it to the end of the list.
Another method of a `list` (and `dict`) is `pop`, which takes an index (or a key) as an argument.
The method `pop` returns the element that is indexed and removes it from the list.

Remember the `better_phonebook` we defined above.

```python
better_phonebook
```

```python
better_phonebook.pop('bill')
```

```python
better_phonebook
```

Finally, one quite useful method of strings is `format()`.
This takes a string with a particular placeholder, `{}` and fills that placeholder with a value provided as an argument.
The placeholder may contain formatting instructions.
For example, suppose that we want to round the approximation of $e$ we obtained above to 3 decimal digits.
We can do so in the following way, using the `{:.3f}` formatter.

```python
print( "We got something like {:.3f}.".format(results[-1]) )
```

### Numpy

[NumPy](https://www.numpy.org/) is a package for Python.
At its core, it defines one thing: N-dimensional arrays.
To better understand, vectors and matrices are 1- and 2-dimensional arrays.
On top of this, NumPy also provides many basic mathematical functions, such as `log()`, `exp()`, `sin()` and so on, together with common mathematical constant like `e`, `pi`, `inf`, `nan` (more below on this one).
The advantage of using NumPy lies in the fact that all operations are optimized to run as fast as possible.

The basic data type NumPy provides is the [`ndarray`](https://docs.scipy.org/doc/numpy/reference/arrays.ndarray.html).
This has several attributes and methods that are useful.
Examples of attributes include: `shape`, `ndim`, and `size`.
Some methods are: `reshape`, `transpose` (shorthanded, `T`), `squeeze` and `min`, `mean`, `argmin`.

Before we dive in some examples, it is useful to clarify something about dimensions.
We typically think of matrices as "rectangles" with two _axes_: rows and columns.
When it comes to vectors, we are normally careful specifying whether it is a row vector or a column vector.
However, in some instances it is not necessary to specify the orientation of a vector (think of the inner product, when we denote it with $\langle x, y \rangle$).
In NumPy, we can have three distinct ways to write a vector:
- one-dimensional vectors, with _no_ orientation;
- two-dimensional row vectors, where the first dimension has size 1; and
- two-dimensional column vectors, where the second dimension has size 1.

This is going to be clear below, but it is important to grasp the existence of vectors whose orientation is irrelevant.
If you are a Matlab user, this is a novelty for you.

Let us start with defining some vectors.

```python
f = np.array([1, 2, 3])
f
```

We can inspect the _shape_ of the vector `f`.

```python
f.shape
```

The shape attribute always returns a `tuple` (can you say why not a `list`?).
There are as many elements in the tuple as the number of dimensions of the arrays (`f.ndim`).
In this case, there is only one element in `f.shape`, which means that orientation of `f` is undefined (neither a row, nor a column vector).

We can access elements of a NumPy array by using square brackets, like we did for lists and tuples.

```python
f[1]
```

We can explicitly give an orientation to a vector.

```python
fc = f.reshape((-1, 1))
fc
```

The `reshape` method takes a `tuple` as argument, with as many elements as dimensions we require.
Here, the value `(-1, 1)` tells NumPy that we want to reorganize `f` such that it has one column and as many rows as require to exhaust the original number of elements.
Obviously, when we reshape, we must be careful that the new require shape allows for exactly as many elements as the original shape (i.e., we had $n$ elements, we gotta have $n$ elements after reshaping).

```python
f.size == fc.size
```

Defining matrices is an obvious extension to the example above.

```python
A = np.array([[  2,   3,   4],
              [200, 150, -10], 
              [ -5,  -2,  90]], dtype=float)
A
```

Note that with the `dtype` argument we can enforce a data type on the elements of the matrix we are creating.
Without `dtype=float`, the matrix `A` would be a matrix of integers, not of floating-point numbers.

Again, we can access elements of a matrix with square brackets.

```python
A[1, 0]
```

A particular value that NumPy defines is the Not-a-Number ([`nan`](https://docs.scipy.org/doc/numpy/reference/constants.html#numpy.NAN)).
This is commonly used to signify that a certain number is not available.
In this sense, it impedes certain operations.
Note that `nan` is _propagated_ by almost all functions applied to it.

```python
np.nan * 2
```

Normally we do not specify matrices "by hand".
Instead, we rely on some functions that generate appropriate arrays.
For example, suppose that we want to evaluate the function $f : X \to Y$, where $X = [0,2]$ and $f(x) = \sqrt{x}$.
While we cannot represent any dense set in a computer, we can discretize it in a small number of _lin_early _space_d points.

```python
X = np.linspace( 0, 2, num=10+1)  # define domain
Y = np.sqrt(X)                    # define image
np.block([X.reshape((-1,1)), Y.reshape((-1,1))])  # collect in two-column matrix
```

We can also create random arrays.

```python
B = np.random.normal(loc=3,       # mean
                     scale=5,     # std. deviation
                     size=(10,))  # no. of elements
B
```

Finally, we can easily create summary statistics about an array.

```python
B.mean()
```

```python
B.std()
```

```python
B.min()
```

```python
B.argmin()
```

```python
B[ B.argmin() ]
```

## Classes (`class`)

Finally, Python provides a way to define objects from scratch.
While an extensive explanation of classes is outside the scope of this tutorial, we are going to see an example.
As we saw above, packages essentially provide new classes and functions.
However, it might well be the case that there is no package that does exactly what we need.
The ability to define our own objects might then come in handy.

Consider the case where you want to define a new object.
This is typically characterized by two things: instance data and methods.
Instance data is the actual data we are going to throw in the object when we create it.
Methods are functions that are tailored to the kind of instance data we think of.

Consider the following example.
Suppose that we want to code up a phonebook.
An entry in a phonebook can be an object.
Its instance data will be the name of the person, the address, the phone number and so on.
The methods will be printing the address on screen or a function that starts the phone call.

Here, we take a look at an object that economists love: a linear regression model.
Such model is characterized by instance data, typically denoted with $y$ and $x$, where the former is the vector of the dependent variable and $x$ are all the independent variables (typically organized in a matrix, a column for each variable).
The methods to a linear regression model may be the following:
- `coefficients`: returning the vector with the estimated regression coefficients;
- `stderrors`: returning the vector of standard errors related to the coefficients;
- `residuals`: returning the vector of residuals;
- `r2`: returning the $R^2$, either normal or adjusted depending on what the user requests.

There might be other methods we can think of (if you are familiar with [R](https://cran.r-project.org/), think of [`summary()`](http://blog.yhat.com/static/img/dino_summary.png), which prints the estimation results in a nice way), but for now we limit ourselves to a basic example.

Operationally:
- We create a class by using the `class` keyword.
- Instance data is assigned by defining a function with the special name `__init__`.
- Each function (method) inside the class must have access to the instance data, which is granted by always specifying the `self` keyword as first argument.

If this sounds unclear, let us give a look at some sample code.

```python
class OLS:

    def __init__(self, y, x, const=True):

        self.y = y
        self.n = y.shape[0]

        if const:
            self.X = np.concatenate((np.ones((self.n, 1)), x), axis=1)
        else:
            self.X = x

        self.k = self.X.shape[1]

        if self.n < 10000:  # leveraging QR decomposition, for precision
            Q, R = la.qr(self.X, mode='economic')
            self.xxi = la.solve(R.T @ R, np.eye(self.k))
        else:               # forget precision, focus on do this fast
            self.xxi = la.solve(self.X.T @ self.X, np.eye(self.k))

        self.beta = self.xxi @ (self.X.T @ self.y)
        self.u = self.y - self.X @ self.beta
        self.sigma2 = (self.u.T @ self.u) / (self.n - self.k)
        self.se_beta = np.sqrt(self.sigma2 * np.diag(self.xxi).reshape((-1,1)))

    def r2(self, adj=False):
        rss = np.asscalar( np.squeeze( self.u.T @ self.u ) )
        tss = np.asscalar( np.squeeze( self.y.T @ self.y ) )
        r2 = 1 - rss/tss

        if not adj:
            return r2
        else:
            return 1 - (1-r2) * ((self.n - 1) / (self.n - self.k))
        
    @property
    def coefficients(self):
        return self.beta.reshape((-1,))
    
    @property
    def stderrors(self):
        return self.se_beta.reshape((-1,))
    
    @property
    def tstats(self):
        return self.beta.reshape((-1,)) / self.se_beta.reshape((-1,))
    
    @property
    def pvalues(self):  # two-sided significance test
        tstats = np.abs( self.beta.reshape((-1,)) / self.se_beta.reshape((-1,)) )
        return ( 1.0 - st.norm.cdf(tstats) ) * 2
    
    @property
    def residuals(self):
        return self.u
```

Let us test this new class with some random data.

```python
# Creating random data (x, y), where y is approximately a linear function of x
n = 100
xLo = 0
xHi = 2
x = np.random.uniform(low=xLo, high=xHi, size=(n,1))
b = np.array([5, -2], dtype=float)
y = b[0] + b[1] * x + np.random.normal(size=(n,1))
    
mdl = OLS(y, x, const=True)
```

```python
mdl.stderrors
```

```python
mdl.tstats
```
